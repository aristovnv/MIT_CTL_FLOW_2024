```python
import pandas as pd
import numpy as np
import sys
sys.path.append("../Coding")
import Constants as con
```

```python
df = pd.read_csv(f"{con.pathData}/ResultedVesselOnBerth.csv")
```

```python

df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

# Sort the DataFrame by 'id' and 'start_time'
df.sort_values(by=['IMO', 'start_time'], inplace=True)

# Step 2: Add a row number per id
df['row_number'] = df.groupby('IMO').cumcount()
df['row_number_match'] = df['row_number'] - 1 #.shift(1)
# Step 3: LEFT join DataFrame to itself on 'id' and 'row_number' with suffix _2
merged_df = pd.merge(df, df, how='left', left_on=['IMO', 'row_number'], right_on=['IMO', 'row_number_match'], suffixes=('', '_2'))

# Step 4: Add column 'diff1_2' as the difference between 'start_time_2' and 'end_time'
merged_df['diff1_2'] = merged_df['start_time_2'] - merged_df['end_time']

# Step 5: Remove rows with 'diff1_2' greater than 7 days
merged_df = merged_df[merged_df['diff1_2'] <= pd.Timedelta(days=7)]
merged_df = merged_df[merged_df['diff1_2'] > pd.Timedelta(days=0)]
merged_df2 = merged_df.copy()
i=2
while True:
    
    merged_df2 = pd.merge(merged_df2, df, how='left', left_on=['IMO', f'row_number_{i}'], right_on=['IMO', 'row_number_match'], suffixes=('', f'_{i+1}'), indicator=True)

    merged_df2[f'diff{i}_{i+1}'] = merged_df2[f'start_time_{i+1}'] - merged_df2[f'end_time_{i}']
    #merged_df2 = merged_df2[merged_df2[f'diff{i}_{i+1}'] <= pd.Timedelta(days=7)]
    merged_df2=merged_df2[merged_df2[(merged_df2[f'diff{i}_{i+1}'].isnull()) | ((merged_df2[f'diff{i}_{i+1}'] <= pd.Timedelta(days=7))&(merged_df2[f'diff{i}_{i+1}'] > pd.Timedelta(days=0)))]]
    #merged_df2 = merged_df2[merged_df2[f'_merge'] == 'left_only']
    merged_df2.drop('_merge', axis=1, inplace=True)
    i += 1
    print(i)
    if len(merged_df2[~merged_df2[f'start_time_{i}'].isna()]) == 0  or i ==15:
        break
```
