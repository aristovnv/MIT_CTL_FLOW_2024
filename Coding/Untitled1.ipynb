```python
ig_folder = f"/home/gridsan/{con.userName}/flow_shared/data/importGenius/Import_"

port_lists = ['Baltimore', 'Charleston', 'NY', 'Phily','Savannah', 'smallPorts','Boston', 'Chester', 'Norfolk','Portland','Wilmington']

#k = port_lists[0]

for k in port_lists:
    make_file(k)
```

```python
def make_file(k):
    my_folder = f"{ig_folder}{k}"
    files = os.listdir(my_folder)
    frames = []
    for f in files:
        #print(f)
            my_file = my_folder + "/" + f
            print(f"processing file {f}")
            df = pd.read_csv(my_file, error_bad_lines=False, index_col=False )
            df = df[df['CONTAINER NUMBER'] != 'NC']
            df_copy = df[['VESSEL NAME','VOYAGE NUMBER','MASTER B/L', 'NO. OF CONTAINERS', 'FOREIGN PORT','US PORT', 'ARRIVAL DATE']]
            df_res = df_copy[df_copy['MASTER B/L']==' ']
            df_res['ARRIVAL DATE'] = pd.to_datetime(df['ARRIVAL DATE'])

            # Extract year and month into separate columns
            df_res['Year'] = df_res['ARRIVAL DATE'].dt.year
            df_res['Month'] = df_res['ARRIVAL DATE'].dt.month

            grouped_df = df_res.groupby(['Year', 'Month', 'VESSEL NAME','VOYAGE NUMBER', 'FOREIGN PORT', 'US PORT']).agg({'NO. OF CONTAINERS': 'sum', 'ARRIVAL DATE': ['min', 'max']})
            grouped_df.columns = ['_'.join(col).strip() for col in grouped_df.columns.values]
            grouped_df.reset_index(inplace=True)
            frames.append(grouped_df)
    resulted_df = pd.concat(frames, ignore_index=True)
    resulted_df.rename(columns={'NO. OF CONTAINERS_sum': 'NO. OF CONTAINERS', 'ARRIVAL DATE_min': 'MIN ARRIVAL DATE', 'ARRIVAL DATE_max': 'MAX ARRIVAL DATE'}, inplace=True)
    resulted_df.info()
    grouped_df = resulted_df.groupby(['Year', 'Month', 'VESSEL NAME','VOYAGE NUMBER', 'FOREIGN PORT', 'US PORT']).agg({'NO. OF CONTAINERS': 'sum', 'MIN ARRIVAL DATE': 'min', 'MAX ARRIVAL DATE': 'max'})
    grouped_df.reset_index(inplace=True)
    save_folder = f"/home/gridsan/{con.userName}/Coding/importGenius/"
    grouped_df.to_csv(f"{save_folder}{k}.csv")
```
